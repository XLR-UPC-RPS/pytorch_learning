## 1.dataloader = DataLoader(dataset, batch_size=4, shuffle=True)其中的shuffle是什么？
shuffle是一个布尔类型（True/False）参数，它的作用是：控制在每个训练epoch开始前，是否对整个数据集的样本顺序进行随机打乱。

## 2.transforms.Normalize(mean=[0.485,0.456,0.406],std=[0.229,0.224,0.225])为什么标准化是这几个值？
针对 ImageNet 数据集的 RGB 三通道图像，统计计算得到的全局均值和标准差，这组值已经成为计算机视觉领域处理 RGB 图像的默认标准之一，尤其是在使用预训练模型的场景下。
## 3.
**stride（步幅）** 是卷积层的核心参数之一，指卷积核在输入特征图上沿水平（宽）和垂直（高）方向滑动的「步长」
**padding（填充）** 是指在输入特征图的四周填充一定数量的「零值像素（零填充）」
对于方形输入、方形卷积核的场景，输出特征图的高 / 宽可以用以下公式精准计算 ：
$$
outputshape = [\frac{inputshape - kernel + 2 * padding}{stride}] + 1
$$
其中，$[·]$表示向下取整
## 4.池化层作用
**缩小尺寸**
让电脑算得更快特征图越大，计算越慢、参数越多。池化直接把宽高缩小（最常见是变成原来的 1/2），大幅减少计算量。
**去掉小噪点、小碎细节**
比如手写数字里的小墨点、小划痕，池化会直接忽略，只保留数字的整体轮廓。

让模型不怕 “图片稍微挪一点”（专业叫：平移不变性）数字往左 / 右偏一点点，池化后结果几乎一样，模型更稳、更通用。

减少参数，防止过拟合特征变小 → 后面全连接层参数变少 → 模型不会死记训练集，泛化更好。
### 池化层的分类：
**最大池化 Max Pooling：**  窗口里只选【最大的那个数】，最突出、最强、最明显的特征，抓特征能力强，抗噪声好
**平均池化 Avg Pooling：**  窗口里算【平均值】，保留整体、平滑、平均的信息，会把强特征 “冲淡”，抓重点不如最大池化，很少用在中间层，一般用在模型最后一层做平滑
**全局平均池化 GAP**: 把整张特征图直接算一个平均值，不用窗口滑，直接压成 1 个值,彻底去掉全连接层，参数极少、极难过拟合，现代 CNN 大量用
